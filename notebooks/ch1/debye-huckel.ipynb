# Import

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Calculation

```python
N = 6.0233*10**23.
c_0=10*10**(-3.)*N*10**3.
epsilon=8.85*10**(-12.)
z=1
e=1.6*10**(-19.)
nm=10**(-9.)
phi=-25*10**(-3.)
k=1.38*10**(-23.)
T=300

phi_0=(e*phi)/(k*T)
lambda_D = np.sqrt((80*epsilon*k*T)/(c_0*e**2.))
```

```python
print lambda_D
```

# Plot

```python
ys = np.linspace(0,25*nm,1000.)
c_pluses = np.array([c_0*np.exp(-1*phi_0*np.exp(-y/lambda_D)) for y in ys])
c_minuses = np.array([c_0*np.exp(1*phi_0*np.exp(-y/lambda_D)) for y in ys])
```

```python
blue = np.array([58/255.,58/255.,249/255.])
red = np.array([249/255.,58/255.,58/255.])
purple = np.array([225/255.,58/255.,249/255.])
yellow = np.array([237/255.,223/255.,61/255.])
```

```python
fig = plt.figure(figsize = (8.5, 8.5*(3./4)))


plt.plot(ys/lambda_D, c_pluses/c_0, c = blue, lw = 3, label = r'$c_{+}$', zorder = 10)
plt.plot(ys/lambda_D, c_minuses/c_0, c = red, lw = 3, label = r'$c_{-}$', zorder = 10)
plt.plot(ys/lambda_D, (c_pluses+c_minuses)/(2.*c_0), lw = 3, ls = '--', c = purple, label = r'$\left(c_{+}+c_{-}\right)/2$', zorder = 5)
plt.plot(ys/lambda_D, (c_pluses-c_minuses)/c_0, lw = 3, ls = '--', c = yellow, label = r'$\left(c_{+}-c_{-}\right)$', zorder = 5)
plt.plot([lambda_D/lambda_D, lambda_D/lambda_D],[-100,100], lw = 3, ls = ':', c = 'k', label = r'$\lambda_{D}$')

plt.xlim(0,5)
plt.ylim(0,3)

plt.legend(loc='upper right', fontsize=12)

plt.title(u'DebyeHckel concentrations', size = 12)
plt.tick_params(labelsize=12)

plt.grid()

plt.xlabel(r'Distance from wall $y/\lambda_{D}$', size = 12)
plt.ylabel(r'Concentration $c/c_{0}$', size = 12)

fig.tight_layout()

plt.savefig('concentrations.png', dpi = 100)

plt.show()
```
