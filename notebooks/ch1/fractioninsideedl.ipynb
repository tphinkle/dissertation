# Imports

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

# Calculation

```python
um = 10**(-6.)
nm = 10**(-9.)

lambda_D = 1*nm
```

```python
Ds = [(1+0.1*i)*nm for i in range(0, 500)]
fractions = [np.pi*((D/2.)**2.-((D-lambda_D)/2.)**2.)/(np.pi*((D/2.)**2.)) for D in Ds]
```

# Plot

```python
fig = plt.figure(figsize = (8.5/2, 8.5/2*(3./4)))

plt.plot(np.array(Ds)/nm, fractions, lw = 3, c = 'k')
plt.scatter(40,0.6, lw = 3, marker = 'o', s = 7500, alpha = 0.5, c = 'red')
plt.scatter(40,0.6, lw = 3, marker = 'o', s = 7500, facecolors = 'none')
plt.scatter(40, 0.6, lw = 3, marker = 'o', s = 2000, c = 'white')

plt.gca().annotate('', xy=(40, 0.6), xytext=(33.5, 0.825),
    arrowprops=dict(arrowstyle="<-"), fontsize = 12)

plt.gca().annotate('', xy=(40, 0.75), xytext=(40, 0.9),
    arrowprops=dict(arrowstyle="<->"), fontsize = 12)

plt.text(32.25, 0.875, r'$D$', fontweight = 'bold', ha = 'center', va = 'center', fontsize = 14)
plt.text(40, 0.975, r'$D-\lambda_{D}$', fontweight = 'bold', ha = 'center', va = 'center', fontsize = 14)


plt.xlim(min(Ds)/nm, max(Ds)/nm)
plt.ylim(0, 1.2)

plt.grid()

plt.xlabel(r'Pore diameter/$\lambda_{D}$', size = 12)
plt.ylabel(r'Fractional volume within $\lambda_{D}$', size = 12)
plt.tick_params(labelsize = 12)

fig.tight_layout()
plt.savefig('fractioninsideedl.png', dpi = 100)

plt.show()
```
