# Get velocities

```python
def filter_data(data, sigma = 50):
    filtered_data = np.copy(data)
    filtered_data[:,1] = scipy.ndimage.filters.gaussian_filter(filtered_data[:,1], sigma)
    return filtered_data
```

```python
def get_min_indices(data):
    maxima, minima = rp.get_maxima_minima(data[:,1])
    
    
    return minima
```

# Crop data

```python
def crop_data(data, sigma = 50, offset = 300):

    processed_data = np.copy(data)
    processed_data[:,1] = scipy.ndimage.filters.gaussian_filter(processed_data[:,1], sigma)
    maxima, minima = rp.get_maxima_minima(processed_data[:,1])
    
    
    return np.copy(data[minima[0]-offset:minima[-1]+offset,:])
```
